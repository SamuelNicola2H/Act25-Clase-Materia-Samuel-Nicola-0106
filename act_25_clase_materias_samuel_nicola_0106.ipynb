```python
class Materia:
    def __init__(self, nombre, clave, creditos):
        # Atributos: Datos referentes a la materia
        self.nombre = nombre
        self.clave = clave
        self.creditos = creditos

    # Funciones: Acciones que realiza la materia
    def Leerdatos(self):
        print(f"\n> Acción: El sistema está leyendo los registros de la materia {self.nombre}.")

    def Mostrardatos(self):
        print(f"> Acción: Desplegando información: {self.nombre} (Clave: {self.clave}) con un valor de {self.creditos} créditos.")

    def ValidarRegistro(self):
        # Tercera función para cumplir con los requerimientos
        print(f"> Acción: Validando que la clave {self.clave} sea correcta en el plan de estudios.")

def crear_materia(numero):
    print(f"\n--- DATOS DE LA MATERIA {numero} ---")
    nom = input("Nombre de la materia: ")
    cla = input("Clave de la materia: ")
    cre = input("Número de créditos: ")

    # Crear objeto (Instancia)
    objeto = Materia(nom, cla, cre)

    # Ejecutar funciones (Acciones)
    objeto.Leerdatos()
    objeto.Mostrardatos()
    objeto.ValidarRegistro()
    return objeto

# --- Ejecución del Programa ---

# Se crean los 2 objetos solicitados
Materia1 = crear_materia(1)
Materia2 = crear_materia(2)

print("\n--- PROCESO DE REGISTRO ACADÉMICO FINALIZADO ---")
print("Programa realizado por Samuel Nicola NL = 40")
```
